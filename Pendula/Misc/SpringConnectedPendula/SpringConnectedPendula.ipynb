```python
from sympy.physics.mechanics import *
import sympy as sp
mechanics_printing(pretty_print=True)
```

```python
m1,m2,l0,k,l1,l2 = sp.symbols(r'm_1,m_2,l_0,k,l_1,l_2')
t,g = sp.symbols('t,g')
v1, v2, p1, p2 = dynamicsymbols(r'\theta_1 \theta_2 \phi_1 \phi_2')
dv1, dv2, dp1, dp2 = dynamicsymbols(r'\theta_1 \theta_2 \phi_1 \phi_2', 1)
```

```python
x1 = l1*sp.sin(p1)*sp.cos(v1)
y1 = l1*sp.sin(p1)*sp.sin(v1)
z1 = l1*sp.cos(p1)

x2 = x1 + l2*sp.sin(p2)*sp.cos(v2)
y2 = y1 + l2*sp.sin(p2)*sp.sin(v2)
z2 = z1 + l2*sp.cos(p2)

dx1 = x1.diff(t)
dy1 = y1.diff(t)
dz1 = z1.diff(t)

dx2 = x2.diff(t)
dy2 = y2.diff(t)
dz2 = z2.diff(t)
```

```python
V = m1*g*y1+m2*g*y2+sp.Rational(1,2)*k*(sp.sqrt((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)-l0)**2
T = sp.Rational(1,2)*m1*(dx1**2+dy1**2+dz1**2)+sp.Rational(1,2)*m2*(dx2**2+dy2**2+dz2**2)
L = sp.simplify(T - V)
```

```python
LM = LagrangesMethod(L, [v1, v2, p1, p2])
```

```python
soln = LM.form_lagranges_equations()
```

```python
soln
```

```python
solvedsoln = sp.solve((sp.Eq(soln[0]),sp.Eq(soln[1]),sp.Eq(soln[2]),sp.Eq(soln[3])),(v1.diff(t,t),v2.diff(t,t),p1.diff(t,t),p2.diff(t,t)))
```
